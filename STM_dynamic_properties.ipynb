**With relaxation length - Yaw natural frequency**

```python
# Effects of rear axle relaxation length
sigmaf = 0.3
sigmar = 0.3
kN = np.sqrt(Jz/(m*lf*(l-lf)))
V = np.linspace(5, 25, num=100)
for sigmar in [0,0.1, 0.3, 0.9]:
    wn = np.sqrt(kf*kr*l*l/(Jz*m*np.square(V)) + (kr*(l-lf) - kf *lf)/Jz)*np.sqrt(1 + 0.5*sigmaf*kf*l/(m*(l-lf)*kN*np.square(V)) + 0.5*sigmar*kr*l/(m*lf*kN*np.square(V)))
    sigma = 0.5*((kf*lf*lf + kr*(l-lf)*(l-lf))/(Jz*V) + (kf + kr)/(m*V))+((0.25*sigmaf*kf*l/(m*(l-lf)*np.square(kN)*V))*(1/l + kf*l/(m*(l-lf)*np.square(V))) - (0.25*sigmar*kr*l/(m*(lf)*np.square(kN)*V))*(1/l - kr*l/(m*(lf)*np.square(V))))
    legend = 'sigmar = ' + str(sigmar)
    plt.plot(V, wn/2/np.pi,label=legend)
plt.legend()
```

```python
# Effects of front axle relaxation length
sigmaf = 0.3
sigmar = 0.3
kN = np.sqrt(Jz/(m*lf*(l-lf)))
V = np.linspace(5, 60, num=100)
for sigmaf in [0,0.1, 0.3, 0.9]:
    wn = np.sqrt(kf*kr*l*l/(Jz*m*np.square(V)) + (kr*(l-lf) - kf *lf)/Jz)*np.sqrt(1 + 0.5*sigmaf*kf*l/(m*(l-lf)*kN*np.square(V)) + 0.5*sigmar*kr*l/(m*lf*kN*np.square(V)))
    sigma = 0.5*((kf*lf*lf + kr*(l-lf)*(l-lf))/(Jz*V) + (kf + kr)/(m*V))+((0.25*sigmaf*kf*l/(m*(l-lf)*np.square(kN)*V))*(1/l + kf*l/(m*(l-lf)*np.square(V))) - (0.25*sigmar*kr*l/(m*(lf)*np.square(kN)*V))*(1/l - kr*l/(m*(lf)*np.square(V))))
    legend = 'sigmaf = ' + str(sigmaf)
    plt.plot(V, wn/2/np.pi,label=legend)
plt.legend()
```
